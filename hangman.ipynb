```python
def playGame(word):
    mistakes = 0
    letters = ""
    guesses = list()
    for i in word: letters+="*"
    while True: 
        print("(Guess) Enter a letter in the word ", letters, end=" ")
        letter = input(" > ")
        if letter in word and letter not in letters and len(letter) == 1:
            for l in range(len(word)): 
                if word[l] == letter: letters = letters[:l]+letter+letters[l+1:]
            print(letters)
        elif (letter in letters or letter in guesses) and len(letter) == 1:
            print("You've already guessed ", letter)
        elif len(letter)>1 and not letter == word:
            print(letter, " is not correct")
            for i in range(len(letter)):
                try:
                    if not letter[i] == word[i]: mistakes+=1
                except:
                    mistakes+=len(letter[i:])
                    break 
            if len(letter)<len(word): mistakes+=len(word)-len(letter)
        elif letter not in word:
            print(letter, " is not in the word")
            mistakes += 1
        if "*" not in letters or letter == word:
            print("The word is ", word, ". You missed ", mistakes, " time(s)")
            break
        elif mistakes == 15:
            print("GAME OVER. The word was ", word)
            break
        guesses.append(letter)
```

```python
import random
words = ["abruptly", "sphynx", "oxidize", "thriftless", "squawk"]
ans = "y"   
while ans == "y":
    word = words[random.randrange(0, len(words)-1)]
    playGame(word)
    ans = input("Do you want to guess another word? Enter y or n> ")
```
